```python
import cv2
import urllib
import numpy as np
from google.colab.patches import cv2_imshow

manFaceCascade = cv2.CascadeClassifier('/content/drive/MyDrive/Dados/haarcascades/haarcascade_frontalface_default.xml')
#catFaceCascade = cv2.CascadeClassifier('https://github.com/cris141187/bibliotecas/blob/main/haarcascade_frontalcatface.xml')
catFaceCascade = cv2.CascadeClassifier('/content/drive/MyDrive/Dados/haarcascades/haarcascade_frontalcatface.xml')

url = 'https://cdn.w600.comps.canstockphoto.com.br/mulher-gato-fotografia-de-stock_csp45095797.jpg'
url_response = urllib.request.urlopen(url)
image_array = np.array(bytearray(url_response.read()), dtype=np.uint8)
image = cv2.imdecode(image_array, -1)

man = manFaceCascade.detectMultiScale(image)
cat = catFaceCascade.detectMultiScale(image)
if (len(man) and len(cat)) == 0:
    print("Nenhuma face encontrada")
else:
    print ("Numero de faces localizadas: " + str(man.shape[0])  + str(cat.shape[0]))
 
    for (x, y, w, h) in man:
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0))
        
    for (x, y, w, h) in cat:
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0))    
 
    cv2_imshow(image)
    while True:
        key = cv2.waitKey(1)
        if key == 27:
            break
 
    cv2.destroyAllWindows()
```
